```python
one = r"/content/drive/MyDrive/projectLBITFW/bookGen"
model = AutoModelForCausalLM.from_pretrained(one)
model_name = "EleutherAI/pythia-1b"
tokenizer = AutoTokenizer.from_pretrained(model_name)
bookPath = "/content/drive/MyDrive/projectLBITFW/chapter3.txt"
```

```python
def cleanResponse(res):
  res = res.replace("<|endoftext|>","")
  while res[-1] == " ":
    res = res[:-1]
  if res[-1] != ".":
    res += "."
  return res
```

```python
def generateBook(bookPath):
  bookFile = open(bookPath,"r").read()
  book = bookFile[-200:]
  starter = prompts[random.randint(0,len(prompts)-1)]
  prompt = book + " " + starter

  input_ids = tokenizer.encode(prompt, return_tensors='pt')
  attention_mask = torch.ones_like(input_ids)

  output_sequence = model.generate(
    input_ids,
    attention_mask = attention_mask,
    pad_token_id = tokenizer.pad_token_id,
    max_length=700)
  response = tokenizer.decode(output_sequence[0])


  response = cleanResponse(response)
  response = response[len(book):]
  book = open(bookPath,"a+")
  book.write(response)
  book.close()
```

```python
while True:
  generateBook(bookPath)
```
